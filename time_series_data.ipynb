# Post-seismic behaviour of the 2011 Mw 9.0 Tohoku-oki earthquake.

The project couples the capabilities of GNSS for enhanced geodynamics and geophysics research. Central to this research is the Tohoku earthquake of 2011 and its consequential tsunami. By using GPS time series data, the project aims to decode the movements and stress transfers of tectonic plates. Furthermore, it builds on Mavrommatis et. all (2011)'s findings, focusing on the deformations due to seismic activities. The ultimate objective is to segregate clear geophysical indications from noise in the GNSS data, which would aid in predicting and understanding geological phenomena better which is the post-seismic behaviour of the 2011 Mw 9.0 Tohoku-oki earthquake. Libraries like `numpy`, `scipy`, `matplotlib`, `statsmodels`, `pytorch` and `scikit-learn` will be used for this project.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import numpy as np


#add the path to the data file

gps_station_203= './Data/950203.tenv3.txt'


# Since the data is in a string format, we'll convert it to a StringIO object to read it with pandas
data_frame = StringIO(gps_station_203)

# If data is separated by whitespace
data_frame = pd.read_csv(gps_station_203, header=0, delim_whitespace=True)

data_frame
```

```python
# plot the East data
plot = plt.plot(data_frame['yyyy.yyyy'], data_frame['__east(m)_______']) 
```
